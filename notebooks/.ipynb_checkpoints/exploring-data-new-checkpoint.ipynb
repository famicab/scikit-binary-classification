```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datasets/Leads.csv')
df.head().T
```

```python
replacer = lambda str: str.lower().str.replace(' ', '_')
df.columns = replacer(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)
df.head().T
```

```python
missing_values = df.isna().sum().to_frame(name='Missing Values').sort_values('Missing Values', ascending=False)
missing_values
```

```python
for i in df.columns:
    if df[i].isna().sum()>3000:
        df.drop(i, axis=1, inplace=True)
```
